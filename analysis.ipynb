```python
a7 = pd.read_csv("Logistics Sample Data/US_Supplier.csv")

combo = a7.groupby(['Supplier_ID', 'Supplier_Name', 'City', 'State' ,'Publisher' ]).size().reset_index(name='count')
print(combo[combo['count'] > 1].sort_values('count', ascending=False))
print(f"\nTotal duplicate combos: {combo[combo['count'] > 1].shape[0]}")
print(f"Total unique combos: {combo[combo['count'] == 1].shape[0]}")
```
