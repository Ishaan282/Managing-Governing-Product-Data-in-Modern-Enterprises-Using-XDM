```python
a7 = pd.read_csv("Logistics Sample Data/US_Supplier.csv")

combo = a7.groupby(['Supplier_ID', 'Supplier_Name', 'City', 'State' ,'Publisher' ]).size().reset_index(name='count')
print(combo[combo['count'] > 1].sort_values('count', ascending=False))
print(f"\nTotal duplicate combos: {combo[combo['count'] > 1].shape[0]}")
print(f"Total unique combos: {combo[combo['count'] == 1].shape[0]}")
```

```python
import pandas as pd

a4 = pd.read_csv("Logistics Sample Data/US_Product.csv")
a7 = pd.read_csv("Logistics Sample Data/US_Supplier.csv")

# merge product with supplier on Supplier_ID
merged = a4.merge(a7, on='Supplier_ID', how='left')

# group by Product_ID and count unique suppliers
product_supplier_count = merged.groupby('Product_ID')['Supplier_ID'].nunique().reset_index()
product_supplier_count.columns = ['Product_ID', 'Supplier_Count']

# filter products that have more than 1 supplier
multi_supplier_products = product_supplier_count[product_supplier_count['Supplier_Count'] > 1]

print(f"Total products: {a4['Product_ID'].nunique()}")
print(f"Products with multiple suppliers: {len(multi_supplier_products)}")
print(multi_supplier_products)
```

```python
import pandas as pd

a2 = pd.read_csv("Logistics Sample Data/US_Customer.csv")

# Count digits in each Mobile number
a2["Digit_Count"] = a2["Mobile"].astype(str).str.count(r'\d')

# Print Mobile numbers with their digit count
print(a2[["Mobile", "Digit_Count"]])
```
