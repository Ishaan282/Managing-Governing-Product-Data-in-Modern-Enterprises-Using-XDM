```python
import pandas as pd
a1 = pd.read_csv("Logistics Sample Data/US_Address.csv")
a2 = pd.read_csv("Logistics Sample Data/US_Customer.csv")
a3 = pd.read_csv("Logistics Sample Data/US_Inventory.csv")
a4 = pd.read_csv("Logistics Sample Data/US_Product.csv")
a5 = pd.read_csv("Logistics Sample Data/US_Sales_Order_Line.csv")
a6 = pd.read_csv("Logistics Sample Data/US_Sales_Order.csv")
a7 = pd.read_csv("Logistics Sample Data/US_Supplier.csv")
a8 = pd.read_csv("Logistics Sample Data/US_Warehouse.csv")
```

```python
a1["Address_ID"].nunique() == len(a1)
```

```python
a2["Customer_ID"].nunique() == len(a2)
```

```python
print("US_Address PK:", 
      a1["Address_ID"].nunique() == len(a1),
      " | Nulls:", a1["Address_ID"].isnull().sum())

print("US_Customer PK:", 
      a2["Customer_ID"].nunique() == len(a2),
      " | Nulls:", a2["Customer_ID"].isnull().sum())

print("US_Inventory PK:", 
      a3["Inventory_ID"].nunique() == len(a3),
      " | Nulls:", a3["Inventory_ID"].isnull().sum())

print("US_Product PK:", 
      a4["Product_ID"].nunique() == len(a4),
      " | Nulls:", a4["Product_ID"].isnull().sum())

print("US_Sales_Order_Line PK:", 
      a5["Order_Line_ID"].nunique() == len(a5),
      " | Nulls:", a5["Order_Line_ID"].isnull().sum())

print("US_Sales_Order PK:", 
      a6["Order_ID"].nunique() == len(a6),
      " | Nulls:", a6["Order_ID"].isnull().sum())

print("US_Supplier PK:", 
      a7["Supplier_ID"].nunique() == len(a7),
      " | Nulls:", a7["Supplier_ID"].isnull().sum())

print("US_Warehouse PK:", 
      a8["Warehouse_ID"].nunique() == len(a8),
      " | Nulls:", a8["Warehouse_ID"].isnull().sum())
```

```python
# Get all duplicate Product_IDs
dup_ids = a4[a4.duplicated('Product_ID', keep=False)].sort_values('Product_ID')

# Show them side by side
print(f"Total records with duplicate Product_IDs: {dup_ids.shape[0]}")
print(dup_ids.head(20).to_string())
```

```python
dup_ids = a1[a1.duplicated('Address_ID', keep=False)].sort_values('Address_ID')
print(f"Total records with duplicate Address_IDs: {dup_ids.shape[0]}")
print(dup_ids.head(20).to_string())
```

```python
# Check if same ID has different addresses
combo = a1.groupby(['Address_ID', 'Address_Line1', 'City', 'State', 'Country', 'Postal_Code']).size().reset_index(name='count')
perfect_dups = combo[combo['count'] > 1]
print(f"Perfect duplicates: {perfect_dups.shape[0]}")
```

```python
# Check UK vs USA concern
# Same ID, different Country
country_conflict = a1.groupby('Address_ID')['Country'].nunique().reset_index()
print(f"Same ID, different Country: {country_conflict[country_conflict['Country'] > 1].shape[0]}")
```
