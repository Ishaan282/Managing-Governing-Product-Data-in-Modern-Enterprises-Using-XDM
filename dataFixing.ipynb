```python
import pandas as pd
a3 = pd.read_csv("Logistics Sample Data/US_Inventory.csv")
a3.info()
```

```python
a3['Inventory_ID'] = a3['Inventory_ID'].str.extract(r'(\d+)').astype(int)
a3['Inventory_ID'] = 'IUS' + a3['Inventory_ID'].astype(str).str.zfill(4)
a3.to_csv("US_Inventory_fixed.csv", index=False)
print("Step 1 done!")
print(a3['Inventory_ID'].head(10))
```
